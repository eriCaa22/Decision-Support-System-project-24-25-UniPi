```python
import pandas as pd

def find_duplicates(file):
    # Carica il file CSV in un DataFrame
    df = pd.read_csv(file, encoding='utf-8')
    
    # Identifica le righe duplicate basandoti su PERSON_ID e CRASH_UNIT_ID
    duplicate_rows = df[df.duplicated(subset=['PERSON_ID', 'CRASH_UNIT_ID'], keep=False)]
    
    return duplicate_rows  # Restituisci le righe duplicate come DataFrame

# Esempio di utilizzo
duplicates_df = find_duplicates('C:/Users/al797/Documenti/GitHub/LDS-Project-24-25/CSV/merged.csv')

# Stampa il DataFrame con i duplicati (opzionale)
duplicates_df
```
