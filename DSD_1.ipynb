Se in un dataset sugli incidenti stradali le colonne CRASH_DATE e DATE_POLICE_NOTIFIED coincidono (cioè indicano lo stesso momento) ma il REPORT_TYPE è indicato come "NOT ON SCENE", potrebbe significare che:

L'incidente è stato segnalato immediatamente: Anche se la polizia o un operatore non erano fisicamente sul luogo dell'incidente al momento della segnalazione, qualcuno ha notificato l'incidente alla stessa ora in cui è avvenuto (ad esempio, tramite telefono o sistema di notifica automatica).

Rapporto registrato in remoto: La segnalazione potrebbe essere stata fatta telefonicamente o online, con i dettagli dell'incidente riportati in tempo reale. Questo può accadere in situazioni in cui non è possibile inviare immediatamente qualcuno sul posto (ad esempio, incidenti minori o situazioni in cui non ci sono feriti gravi).

Sistema di registrazione automatizzato: In alcuni casi, soprattutto se ci sono tecnologie avanzate di rilevamento, come nei veicoli moderni dotati di sistemi di notifica automatica (come eCall in Europa), l'incidente può essere segnalato automaticamente alla stessa ora in cui si verifica, ma non viene considerato "on scene" poiché nessuno è fisicamente presente sul posto
