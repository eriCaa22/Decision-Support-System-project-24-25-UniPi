```python

# Caricare il dataset da un file CSV
file_path = 'LDS24 - Data/Crashes.csv'
crash = pd.read_csv(file_path)

# Visualizzare le prime righe del dataset
print(crash.head())
```

```python
for col in crash.columns:
    if crash[col].isnull().sum()>0:
         print(col, crash[col].isnull().sum())
```
